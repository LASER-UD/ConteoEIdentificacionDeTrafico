```python
#Importamos tensorflow y Keras, primero instalamos tensorflow-gpu en caso de que no esté
import tensorflow
from tensorflow import keras
```

```python
#Ahora crearemos los train generator para cada clase
#Ahora normalizamos los canales de RGB de entrenamiento
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(
    rescale=1/255,
    validation_split = 0.3)

#Creamos el generador de imágenes de entrenamiento cuyo tamaño será de 200x200 y estará en batches de 32 con 2 clases
train_generator = train_datagen.flow_from_directory(
    directory='Vehiculos',
    target_size=(200, 200),
    class_mode='binary',
    subset='training') # set as training data

validation_generator = train_datagen.flow_from_directory(
    directory='Vehiculos', # same directory as training data
    target_size=(200, 200),
    class_mode='binary',
    subset='validation') # set as validation data
```

```python
#Creamos el modelo
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
model = Sequential()
model.add(Conv2D(32,(7,7),input_shape=(200,200,3),activation='relu',kernel_initializer='random_normal',bias_initializer='zeros'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(1,activation='sigmoid',kernel_initializer='random_normal',bias_initializer='zeros'))
model.compile(loss='binary_crossentropy',metrics=['acc'])
model.summary()
```

```python
!pip install pygad
import pygad.kerasga
import numpy
import pygad
```

```python
def fitness_func(solution, sol_idx):
    global keras_ga, model, train_generator

    model_weights_matrix = pygad.kerasga.model_weights_as_matrix(model=model,
                                                                 weights_vector=solution)

    model.set_weights(weights=model_weights_matrix)
    predictions = model.evaluate(train_generator, verbose = 0)
    
    solution_fitness = 1.0 / (predictions[0] + 0.000000001)

    return solution_fitness
```

```python
def callback_generation(ga_instance):
    print("Generation = {generation}".format(generation=ga_instance.generations_completed))
    print("Fitness    = {fitness}".format(fitness=ga_instance.best_solution()[1]))
```

```python
# Creamos la instancia de kerasGA con el número de soluciones de la población inicial
keras_ga = pygad.kerasga.KerasGA(model=model, num_solutions=10)
```

```python
# Preparamos los parámteos del algoritmo genético
num_generations = 150 # número de generaciones
num_parents_mating = 5 # número de padres que serán seleccionados para crossover
initial_population = keras_ga.population_weights # Población inicial de pesos de las redes neuronales.
parent_selection_type = "rws" # Método de seleccion por ruleta.
crossover_type = "single_point" # Crossover en un punto random.
mutation_type = "random" # Tipo de operador de mutación.
mutation_percent_genes = 20 # Porcentaje de genes para mutar.
keep_parents = -1 # Guardar los padres para la población siguiente
```

```python
# Create an instance of the pygad.GA class
ga_instance = pygad.GA(num_generations=num_generations,
                       num_parents_mating=num_parents_mating,
                       initial_population=initial_population,
                       fitness_func=fitness_func,
                       parent_selection_type=parent_selection_type,
                       crossover_type=crossover_type,
                       mutation_type=mutation_type,
                       mutation_percent_genes=mutation_percent_genes,
                       keep_parents=keep_parents,
                       on_generation=callback_generation)
```

```python
# Start the genetic algorithm evolution.
ga_instance.run()
```

```python
# After the generations complete, some plots are showed that summarize how the outputs/fitness values evolve over generations.
ga_instance.plot_result(title="PyGAD & Keras - Iteration vs. Fitness", linewidth=4)

# Returning the details of the best solution.
solution, solution_fitness, solution_idx = ga_instance.best_solution()
print("Fitness value of the best solution = {solution_fitness}".format(solution_fitness=solution_fitness))
print("Index of the best solution : {solution_idx}".format(solution_idx=solution_idx))

# Fetch the parameters of the best solution.
best_solution_weights = pygad.kerasga.model_weights_as_matrix(model=model,
                                                              weights_vector=solution)
model.set_weights(best_solution_weights)
predictions = model.predict(train_generator)
# print("Predictions : \n", predictions)

loss = model.evaluate(train_generator)
print("Categorical Crossentropy : ", loss[0])
```
